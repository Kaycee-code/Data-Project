ANALYTIC QUESTION 1:

Who were the passengers on the Titanic? (Gender, Ages, Passenger Class,..etc)

INSIGHT 1:
Let's start by visualizing the Passenger boarding classes with respect to gender:

```python
# use a countplot to visualize the passengers with respect to their gender and boarding class:

plot = sns.countplot(x='Pclass', data=titanic_df, hue='Sex', palette='Reds')
plot.set(xlabel='Passenger Class', ylabel='Number of Passengers', title='Number of Passengers in Each Class')
```

FINDINGS:

1. we have approximately 350 male passengers in the 3rd class
2. The 2nd class had the least number of passengers (male and female).
3. Overall, passengers favoured boarding the 3rd class of the titanic which we can assume to attract a cheaper fare.

INSIGHT II:

Let's further classify the passengers as adult and children(less than 16 years)

```python
#create a function to check if a passenger is either a child or an adult: 

def adult_child(passenger):
    age = passenger
    if age <16:
        return 'Child'
    else:
        return 'Adult'
```

```python
#Create a new column by applying the above function on titanic_df
titanic_df['Person'] = titanic_df['Age'].apply(adult_child)
```

```python
#Test to confirm that the above function was implemented
titanic_df.head(10)
```

```python
#generate two visuals, side by side showing the numerical and percentage values of adults and children that boarded the titanic:

plt.figure(figsize=(12, 5))

plt.subplot(1,2,1)
chart = sns.countplot(x='Person', data= titanic_df, hue = 'Sex', palette = 'Blues')
chart.set(xlabel = 'Passenger category', ylabel='Number of passengers', title = "Numerical Frequency of Passenger's Age Category")

plt.subplot(1,2,2)
chart = sns.countplot(x='Person', data= titanic_df, hue = 'Sex', palette = 'Blues', stat='percent')
chart.set(xlabel = 'Passenger category', ylabel='Percentage', title = "Percentage of Passenger's Age Category")
```

```python
#To verify the actual number of children (below 16 years):

titanic_df.Person.value_counts().sort_values()
```

FINDINGS:

1. There were more adults than children on board.
2. 60% of passengers on board were male adults (approximately 520 male adults).
3. There were less than 100 children on board.

INSIGHT 3:

Let's shade more light on the age distribution of the passengers aboard the Titanic

Using a Histogram and a KDE plot:

```python
#using a histogram to examine the age distribution:

plt.figure(figsize=(18, 5))

titanic_df['Age'].hist(bins=70)

#To have a deeper look at the age distribution of male and female passengers respectively, we will deploy the KDE plot

sns.displot(data=titanic_df, x='Age', hue='Sex', kind='kde', aspect=2, fill=True)
oldest = titanic_df['Age'].max()
plt.xlim(0, oldest)

```

FINDINGS:

1. The highest age bracket are passengers in their late 20s to early 30s
2. There were more 24-year olds on board than any other age.
3. The number of children between age 0-13 experienced a dip as the age value increased.
4. The youngest and oldest passengers were less than 1year and 80years old respectively. 

ANALYTIC QUESTION 2:

What decks were the passengers on how many were on each deck?

Let's get started by finding out how many decks were housing passengers

```python
#Let's examine the Cabin column of the titanic dataframe again:
titanic_df.head()
```

```python
#Remember that there are only 204 non-null entries in the Cabin column
#get rid of the null values in the 'Cabin' Column

deck = titanic_df.Cabin.dropna()
```

```python
#lets examine the resulting series, deck:

deck.head()
```

```python
#Lets create an empty list called levels and use it to store the first character of all entries in deck:
levels = []
for level in deck:
    levels.append(level[0])

#let's create a dataframe called cabin_df from the resulting list, levels:
cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']

cabin_df.Cabin.unique() #To see the number of unique decks in the Titanic
```

```python
#Lets generate a bar chart showing the number of passengers in each deck:

a = cabin_df['Cabin'].sort_values() #to arrange the decks alphabetically
chart = sns.countplot(data = cabin_df, x=a, hue=a, palette = 'summer')
chart.set(xlabel='Decks', ylabel='Number of Passengers', title='Passengers in various Decks')
```

```python
#To further examine the dataframe cabin_df 

cabin_df.describe()
```

FINDINGS:

1. There were 59 passengers confirmed to be on Deck C on the titanic.
2. Deck G appears to be the deck with the least number of passengers.
3. Much inferences cannot be made from this column as 75% of the column entries were null values.

ANALYTIC QUESTION 3:

Which locations did the passengers board from?


Hint: The "Embarked" column gives the details of where each passenger embarked on the journey from 
and we will dig in to see what boarding class is predominant in each location.

```python
#First, lets replace the Q, S, C in the Embarked column with the full city name.

titanic_df.Embarked = titanic_df.Embarked.replace({'S':'Southampton', 'Q':'Queenstown', 'C': 'Cherbourg'})
titanic_df.head()
```

```python
#Next, lets visualize the number of passengers that boarded at the three ports with respect to their passenger class
#using a countplot and a pie chart

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
chart = sns.countplot(data = titanic_df, x='Embarked', hue='Pclass', order=['Queenstown', 'Cherbourg', 'Southampton'], palette = 'summer')
chart.set(xlabel='Locations', ylabel='Number of passengers', title='Passengers vs Locations')

plt.subplot(1,2,2)
titanic_df['Embarked'].value_counts().plot.pie(
    autopct = '%1.1f%%',
    figsize = (12, 5),
    title = 'Passengers from Various Locations',
    colors = ['skyblue', 'lightcoral', 'lightgreen'],
    explode = (0.1, 0.1, 0), # Explode the 1st slice
    shadow = True,
    ylabel=''
)
plt.show()
```

FINDINGS:

1. Over 72% of the entire passengers boarded from Southampton
1. Approximately 350 passengers from Southampton boarded with 3rd class tickets.
2. Queenstown had the lowest number of passengers, about 9%.
3. A greater fraction of passengers from Cherbourg boarded with first class tickets

ANALYTICAL QUESTION 4:

Who was alone and who was with family? 

How did it affect the chances of survival?

```python
#SibSp = Above 0 means the passenger has sibling(s) on board and 0 means they do not
#Parch = Above 0 means the passenger has parent/child on board and 0 means they do not

#Lets create a new column called family_onboard to check if a passenger had a family member onboard

titanic_df['Family_onboard'] = titanic_df['SibSp'] + titanic_df['Parch']
for number in titanic_df.Family_onboard:
    if number > 0:
        titanic_df['Family_onboard'] = titanic_df.Family_onboard.replace(number, 'With Family')
    else:
        titanic_df['Family_onboard'] = titanic_df.Family_onboard.replace(number, 'Alone')
    
titanic_df.head()
```

```python
#### Let us visualize the number of passengers with family compared to lone passengers with respect to their gender
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
chart = sns.countplot(data= titanic_df, x='Family_onboard', hue = 'Sex', palette = 'summer')
chart.set(xlabel= 'Family Onboard', ylabel= 'Number of passengers', title='Passengers with Family Members according to Gender')

plt.subplot(1,2,2)
titanic_df['Family_onboard'].value_counts().plot.pie(
    autopct ='%1.1f%%', 
    shadow =True, 
    explode =(0.1,0),
    title = 'Passengers with Family Members',
    colors = ['Yellow', 'Green'],
    ylabel='')
```

FINDINGS:

1. Over 400 male passengers boarded the titanic alone
2. Less than 150 female passengers boarded the titanic alone
3. Also, approximately 40% of passengers had a family member on board.

INSIGHT 2:

Still on Analytical Question 4, we need to create a column to show whether a passenger survived the wreck or not.

The Survived column shows 0 and 1 for non-survivors and survivors respectively.

```python
#create a column called survivor and map the entries in the survived column; 1 for Yes and 0 for No

titanic_df['Survivor'] = titanic_df.Survived.map({0:'No', 1:'Yes'})

chart = sns.countplot(data=titanic_df, x='Sex', hue='Survivor', palette= 'Blues')
chart.set(ylabel='No of passengers', title='Survivors and Non-survivors according to Gender')
```

FINDINGS:
1. Over 200 female passengers and approximately 100 male passengers survived the wreck.
2. Over 450 male passengers and less than 100 female passengers lost their lives in the titanic wreck.

INFERENCE:
We can draw an assumption that more effort and resources like life boats were deployed to save female passengers first.


Now let us examine the rate of survival with respect to Passenger's gender and having a family member on board

Did having a family member increase the chances of surviving the crash?:

```python
#Let us utilize three visuals to see how having a family member on board correlated to survival rate:

fig = plt.figure(figsize = (14,4)) #to set the default sizes of the visuals

ax1=fig.add_subplot(131)
chart1 = sns.pointplot(data=titanic_df, x='Family_onboard', y='Survived', ax=ax1)
chart1.set(ylabel='Rate of Survival', xlabel = 'Family Member Onboard', title='Chart 1')

ax2= fig.add_subplot(132)
chart2 = sns.pointplot(data=titanic_df, x='Family_onboard', y= 'Survived', hue='Sex', ax=ax2)
chart2.set(ylabel='Rate of Survival', xlabel = 'Family Member Onboard', title= 'Chart2')

plt.subplot(1,3,3)
chart3 = sns.countplot(data=titanic_df, x='Family_onboard', hue='Survivor', palette= 'Blues')
chart3.set(xlabel= 'Family Member Onboard', ylabel='No of passengers', title='Chart 3')
```

FINDINGS:

1. From chart 1, we see that passengers who had family members had a higher rate of survival
2. Considering gender, chart 2 shows that lone females had higher survival rate than females with family members
3. Again, lone male passengers had a lower survival rate than the males with family members. Infact they had the lowest survival rate.
4. Over 350 lone passengers did not survive the wreck while about 175 passengers with family failed to survive as well.
5. About 180 passengers who had family members survived as compared to 170 lone survivors (approximately)

INFERENCES:
1. Having a family member might have increased the chances of survival
2. Majority of the casualties were Lone male passengers, maybe they tried rescuing their female counterparts.

ANALYTIC QUESTION 5:

Did the youngest and oldest passenger on the titanic survive?

```python
#Extract the youngest and oldest ages from the ages column
youngest = titanic_df.Age.min()
oldest = titanic_df.Age.max()

#using the above variables, extract their corresponding rows from titanic_df 
young_df = titanic_df[titanic_df.Age == youngest]
old_df = titanic_df[titanic_df.Age == oldest]

#create a new df by merging the two extracted DFs above
young_and_old = pd.concat([young_df, old_df], axis=0)
```

```python
#Using a pie chart to generate the visualization:

young_and_old['Survivor'].value_counts().plot.pie(
    ylabel = '',
    title = "Oldest and Youngest Passenger's Survival Status",
    autopct = '%1.1f%%'
)
```

FINDINGS:

1. The youngest and oldest passengers survived the wreck
